**Membaca Dictionary Slang**

```python
alay_dict = pd.read_csv("data_crawling/new_kamusalay.csv", encoding='latin1')
alay_dict = alay_dict.rename(columns={'anakjakartaasikasik': 'original', 
                                     'anak jakarta asyik asyik': 'replacement'})
alay_dict.tail(10)
```

**Membaca Dictionary StopWord**

```python
id_stopword_dict = pd.read_csv("data_crawling/stopwordbahasa.csv")
id_stopword_dict = id_stopword_dict.rename(columns={"ada": 'stopword'})
id_stopword_dict = id_stopword_dict['stopword'].append(pd.Series('ada'), ignore_index=True)
id_stopword_dict = pd.DataFrame({'stopword': id_stopword_dict})
id_stopword_dict.head()
```

**TOKENISASI**

```python
import re
from nltk.tokenize import RegexpTokenizer
regexp = RegexpTokenizer(r'w+|$[0-9]+|\S+')
train_df['Token'] = train_df['Stemming'].apply(regexp.tokenize)
```

```python
train_df['Token'].iloc[200]
```

**Merubah Slang Menjadi Baku**

```python
alay_dict_map = dict(zip(alay_dict['original'], alay_dict['replacement']))
def normalize_alay(text):
    return ' '.join([alay_dict_map[word] if word in alay_dict_map else word for word in text.split(' ')])
```

```python
train_df['EYD'] = train_df['tweet'].apply(normalize_alay)
```

```python
train_df['EYD'].iloc[200]
```

**Remove StopWord**

```python
stopwords_new = pd.DataFrame(['sih','nya', 'iya', 'nih', 'biar', 'tau', 'kayak', 'banget', 're', 'ti', 'ya', 'k', 'p'], columns=['stopword'])
id_stopword_dict = pd.concat([id_stopword_dict,stopwords_new]).reset_index()
id_stopword_dict = pd.DataFrame(id_stopword_dict['stopword'])
```

```python
def remove_stopword(text):
    text = ' '.join(['' if word in id_stopword_dict.stopword.values else word for word in text.split(' ')])
    text = re.sub('  +', ' ', text) # Remove extra spaces
    text = text.strip()
    return text
```

```python
train_df['remove_stopword'] = train_df['EYD'].apply(remove_stopword)
```

```python
train_df['remove_stopword'].iloc[200]
```

**Stemming**

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(text):
    return stemmer.stem(text)
```

```python
train_df['Stemming'] = train_df['remove_stopword'].apply(stemming)
```

**Labelling**

```python
#sentiment analisis dengan polarity menggunakan library textblob
from textblob import TextBlob
def analize_sentiment(tweet):
    analysis = TextBlob(tweet)
    if analysis.sentiment.polarity > 0:
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
```

```python
train_df['label'] = train_df['Stemming'].apply(analize_sentiment)
```
