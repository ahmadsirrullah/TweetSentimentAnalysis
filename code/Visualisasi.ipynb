```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_v2 = pd.read_csv('data_crawling/data_mutakhir.csv')
df_v2
```

```python
df_v2 = df_v2.reset_index()
df_v2
```

```python
import re
from nltk.tokenize import RegexpTokenizer
regexp = RegexpTokenizer(r'w+|$[0-9]+|\S+')
df_v2['Token'] = df_v2['Stemming'].apply(regexp.tokenize)
```

```python
df_pos = df_v2[df_v2['label'] == 1]
df_neg = df_v2[df_v2['label'] == 0]
```

```python
df_pos.count()
```

```python
df_neg.count()
```

```python
df_pos = df_pos.reset_index()
df_neg = df_neg.reset_index()
```

```python
token_pos = []

for i in range(61):
    for j in df_pos['Token'][i]:
        token_pos.append(j)
        
token_pos
```

```python
token_neg = []

for i in range(429):
    for j in df_neg['Token'][i]:
        token_neg.append(j)
        
token_neg
```

```python
from nltk.tokenize import word_tokenize
import nltk
freq_pos = nltk.FreqDist(token_pos)
freq_neg = nltk.FreqDist(token_neg)
print(freq_pos.most_common(10))
print(freq_neg.most_common(10))
```

```python
print(len(token_pos))
print(len(token_neg))
```

```python
for i in range(869):
    if token_pos[i] == 'ya':
        #kata_baru = token_pos[i]+token_pos[i+1]
        del token_pos[i]
        #del token_pos[i+1]
        #print(tokens[i-1],'\n',tokens[i],'\n',tokens[i+1])
        #token_pos.insert(i,kata_baru)
    else:
        continue
```

```python
for i in range(5808):
    if token_neg[i] == 'ya':
        #kata_baru = token_neg[i]+token_neg[i+1]
        del token_neg[i]
        #del token_neg[i+1]
        #print(token_neg[i-1],'\n',token_neg[i],'\n',token_neg[i+1])
        #token_neg.insert(i,kata_baru)
    else:
        continue
```

```python
kata_pos = []
jumlah_kata_pos = []
for key,value in freq_pos.items():
    kata_pos.append(key)
    jumlah_kata_pos.append(value)
```

```python
kata_neg = []
jumlah_kata_neg = []
for key,value in freq_neg.items():
    kata_neg.append(key)
    jumlah_kata_neg.append(value)
```

```python
kata_positif= pd.DataFrame({
    'Kata': kata_pos,
    'Frekuensi': jumlah_kata_pos
})

kata_positif = kata_positif.set_index('Kata')
```

```python
kata_negatif= pd.DataFrame({
    'Kata': kata_neg,
    'Frekuensi': jumlah_kata_neg
})

kata_negatif = kata_negatif.set_index('Kata')
```

```python
kata_positif.sort_values(by = 'Frekuensi').tail(10).plot(kind = 'barh', figsize=(15, 10));
plt.title("Most Common Positive Words", size=17)
plt.show()
```

```python
kata_negatif.sort_values(by = 'Frekuensi').tail(10).plot(kind = 'barh', figsize=(15, 9));
plt.title("Most Common Negative Words", size=17)
plt.yticks(label = kata_negatif.index, fontsize = 10)
plt.show()
```
